```python
# Write a function to return whether an 8-bit binary string has equal no of zeros and ones or not (balanced or unbalanced).
def balanced(string):
    count = 0
    for i in string:
        if i == '1':
            count += 1
    if count == 4:
        return True
    else:
        return False
string = input('Enter an 8-bit binary string:')
print('{} is {} binary string!'.format(string,'balanced' if balanced(string) else 'unbalanced'))
```

```python
# Write a function to take two values as parameters and return ratio of two values.
a, b = [int(input('Enter a number:')),int(input('Enter a number:'))]
def ratio(a,b):
    temp = min(a,b)
    while temp > 1:
        if not(a%temp) and not(b%temp):
            a = a/temp
            b = b/temp
            break
        else:
            temp -= 1
    return str(int(a))+':'+str(int(b))
print('ration of {}:{} is {}'.format(a,b,ratio(a,b)))
```

```python
'''
A dictionary contains brands as keys() and list of MRP and discount percentage as values().
Write a function to return dictionary with brand and final_price.
products = {'brand-i':[MRP-i,discount%-i]............}
'''
def final_price(temp):
    dummy = dict([])
    for i in temp:
        dummy.update({i:int(temp[i][0]*(1-(temp[i][1]/100)))})
    return dummy
products = {'Dell':[75000,20],'Mac':[90000,10],'HP':[60000, 15],'Lenovo':[60000,25],'Acer':[55000,15]}
final = final_price(products)
print('Products:',products)
print('final_price:',final)
```

```python
'''
Write a function to return the list of capital letters in a string.
Hint:  by checking ASCII values or
       i == i. upper ()
'''
def check(string):
    checking = string.lower()
    temp = []
    for i in range(len(string)):
        if string[i] != checking[i]:
            temp.append(string[i])
    return temp
string = input('Enter a string:')
print('capital letters in {} is {}.'.format(string, check(string)))
```

```python
# write a program to take list of integers and convert there signs ie positive to negative and vice-versa.
def convert(nums):
    for i in range(len(nums)):
        nums[i] = -nums[i]
    return nums
numbers = [23, -91, -82, 74, 41, 0, -21, 11, 35, -66, -29]
print('Numbers_list:',numbers)
print('Numbers_converstion:',convert(numbers))
```
