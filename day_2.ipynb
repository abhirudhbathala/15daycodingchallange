```python
# Write a program to find the index of an element in a list.
string=input('Enter the elements of the list by comma-saparated sequence:')
lst=string.strip().split(',')
print('List:',lst)
sample=input("Enter element to find it's index:")
if sample in lst:
    print('The element is found in list at index:',lst.index(sample))
else:
    print('The element is not found in the list!')
```

```python
# Write a program to print the Fibonacci series using recursion?
def fibo(n):
    if n<=0:
        return 0
    elif n==1:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
num=int(input('Enter the number of elements to print in fibonnaci series:'))
fibo_list=[]
for i in range(num):
    fibo_list.append(fibo(i))
print('fibo_series:',fibo_list)
```

```python
# Write a program to find the maximum of three numbers using functions?
def max_three(a,b,c):
    if a>b and a>c:
        return a
    elif b>a and b>c:
        return b
    else:
        return c
a,b,c=[int(input('num_'+i+':')) for i in 'abc']
print('The maximum number of three numbers is '+str(max_three(a,b,c)))
```

```python
# Write a program to find the maximum element in a list without using inbuilt methods?
string=input('Enter sequence of numbers seperated by comma to be stored in list:')
nums=[int(i) for i in string.strip().split(',')]
temp=nums[0]
for i in nums:
    if temp<i:
        temp=i
print('List:',nums)
print('The maximum element of the list is',temp)
```

```python
#  Write a program to print the sum of factorials of two numbers without using inbuilt methods?
def fact(n):
    if n<=1:
        return 1
    else:
        return n*fact(n-1)
a,b=[int(input('Enter a number:')) for i in range(2)]
print('The sum of factorial of {} and {} is'.format(a,b),fact(a)+fact(b))
```

```python
'''
-  Create a class Student.
-  Create an object of Student class.
-  The object should have name, age, marks1, marks2, marks3 (attributes).
-  Print all the attributes.
'''
class Student():
    def __init__(self,name,age,*marks):
        self.name=name
        self.age=age
        self.marks=marks
obj=Student('Ravi',21,98,92,95)
print('Attributes of the object:')
print('Name of student:',obj.name)
print('Age of student:',obj.age)
for i in range(3):
    print('marks{}:'.format(i+1),obj.marks[i])
```

```python
'''
-  Create a method inside the class that returns the sum of marks.
-  Print the sum.
'''
class Student_total(Student):
    def total_marks(self):
        sum=0
        for i in self.marks:
            sum+=i
        return sum
obj=Student_total('Ravi',21,98,92,95)
print('The total marks of {} is {}'.format(obj.name,obj.total_marks()))
```

```python
'''
- Create a method inside the class that returns the average of marks.
-  Print the average.
'''
class Student_average(Student):
    def average(self):
        avg=0
        for i in self.marks:
            avg+=i/3
        return avg
obj=Student_average('Ravi',21,98,92,95)
print('The average marks of {} is {}'.format(obj.name,obj.average()))
```
