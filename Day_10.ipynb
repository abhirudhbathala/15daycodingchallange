```python
'''
Create a class with private variables and methods and access private variables and method.
Syntax of private members in class:
Private variable -> self. __variable = variable
Private method -> def __method(self):
Hint: private variables and methods cannot be called directly. We can access them through public or protected methods.
'''
class Student:
    def __init__(self, name, age, *marks):
        self.__name = name
        self.__age  = age
        self.__marks = marks
    def __total(self):
        temp = 0
        for i in self.__marks:
            temp += i
        return temp
    def __average(self):
        return self.__total()/len(self.__marks)
    def _show(self):
        print('Student_name:',self.__name)
        print('Student_age:',self.__age)
        print('Student_marks:',self.__marks)
        print('Total marks :',self.__total())
        print('Average marks :',self.__average())

obj = Student('Manoj', 19, 95, 93, 97)
obj._show()
```

```python
'''
Print an equilateral triangle pattern using ‘*’.
Hint: Frist three line of code:
        n = int (input (‘Enter number of lines:’))
    	Count = n-1
    	for i in range (1, n*2, 2):
    		statements……
'''
n = int(input('number of lines:'))
count = n-1
for i in range(1,n*2,2):
    string = ' '*count+'*'*i
    print(string)
    count-=1
```

```python
'''
Write a function to take list as parameter and return a dictionary that contains information of how many times each element is repeated in list.
Do without import Counter from collections.
Eg: list = ['a', 'b', 'c', 'd', 'a', 'e', 'a', 'c', 'd', 'e']
    dict = function(list)
    where dict will be {'a': 3, 'b': 1, 'c': 2, 'd': 2, 'e': 2}
'''
temp = ['a','b','c','d','a','e','a','c','d','e']
def fun(items = temp):
    counter = {}
    for i in items:
        if i not in counter.keys():
            counter[i] = 1
        else:
            counter[i]+=1
    return counter
print('Actual_list:',temp)
print('Count of each elements of list:',fun())
```

```python
# write a program to find the character with highest ASCII vlaue in string.
string = input('Enter a string:')
temp = string[0]
for i in string:
    if ord(temp) < ord(i):
        temp = i
print("The character in string with highest ASCII value is '{}'.".format(temp))
```

```python
# Write a program to add three 2-D matrix
A = [[2,4],[8,2]]
B = [[8,1],[4,1]]
C = [[2,1],[3,4]]
matrix = [[None,None],[None,None]]
for i in range(2):
    for j in range(2):
        matrix[i][j] = sum(x[i][j] for x in [A,B,C])
print('Matrix-A:',A)
print('Matrix-B:',B)
print('Matrix-C:',C)
print('Addition of three matrix:',matrix)
```

```python
'''
Write a program to add elements of 3-D matrix.
Eg:  3-D matrix = [[1,1,1],[1,1,1],[1,1,1]]
     sum of elements in matrix = 9
'''
matrix = [[9,8,3],[9,3,1],[3,2,4]]
total = 0
for i in matrix:
    for j in i:
        total += j
print('Sum of elements in matrix:',total)
```
